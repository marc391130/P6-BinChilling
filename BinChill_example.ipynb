```python
def sim_c(c1: set, c2: set, n):
    top = len(set.intersection(c1,c2)) - ((len(c1) * len(c2)) / n)
    bottom = sqrt(len(c1) * len(c2) * ( 1 - (len(c1) / n) ) * (1 - (len(c2) / n)))
    return top / bottom


def print_all_SC(C, n):
    m = -1
    for i, c1 in enumerate(C):
        for j in range(i+1, len(C)):
            c2 = C[j]
            sim = sim_c(c1,c2,n)
            m = max(m, sim)
            print(i,j, sim)
    print(f"\nmax S_c: {m}")
    
def full_score_scg(bin: set, map: dict, G: set):
    scgs = list([scg for item in bin for scg in map.get(item, [])])
    if len(scgs) == 0: return 0.0
    com = len(set(scgs)) / len(G)
    con = len(set([g for g in scgs if scgs.count(scgs) > 1])) / len(set(scgs))
    megp = (len(scgs) - len(set(scgs))) / len(G)
    com, con, megp = com*100, con*100, megp*100
    return com - (con**2) - sqrt(megp)
```

```python
#Handling uncertain items with SCG:
#items: 1, 2, 3, 7, 10

def score_bc(cluster: set, item, cluster_index):
    without = cluster if item not in cluster else {x for x in cluster if x != item}
    with_ = cluster if item in cluster else cluster.union([item])
    score = (S_x.get( (item, cluster_index), 0.0 ) / k )  * ( score_scg(with_) - score_scg(without))
    return score
    

def find_max_scg(item):
    best, ms = -1, None
    for i, c in enumerate(C):
        if item not in c: continue
        s = score_bc(c, item, i)
        print(f"item {item} with cluster {i} score {s}")
        if s < 0: continue
        if ms is None or s > ms:
            best, ms = i, s
    return best

def remove_item_from_all_clusters(item):
    for c in C:
        c : set
        if item not in c: continue
        c.remove(item)

def assign_item_to_cluster(item, cluster_index):
    if cluster_index < 0: 
        print(f"item {item} with no cluster")
        return
    remove_item_from_all_clusters(item)
    c: set = C[cluster_index]
    c.add(item)
    return (item, cluster_index)

def print_assignment(item):
    i, c_i = assign_item_to_cluster(item, find_max_scg(item))
    print(f"item {i} assigned to cluster {c_i}")

print_assignment(1)
print_assignment(2)
print_assignment(3)
print_assignment(7)
print_assignment(10)
print("clusters:")
for c in C:
    print(c)
```

```python
#assignment of uncertain without scgs
#clusters 6, 8, 9, 10

def find_max(item):
    best, ms = -1, None
    for i, c in enumerate(C):
        score = S_x.get( (item, i), 0.0 ) / 3
        if score <= 0.0: continue
        print(f"item {item} with cluster {i} score {score}")
        if 0.5 < score and (ms is None or score > ms):
            best, ms = i, score
    return best

def print_assignment_wo_scg(item):
    i, c_i = assign_item_to_cluster(item, find_max(item))
    print(f"item {i} assigned to cluster {c_i} wo scgs")
    

print_assignment_wo_scg(6)
print_assignment_wo_scg(8)
print_assignment_wo_scg(9)


print("clusters:")
for c in C:
    print(c)
```

```python
#stage 3
#setting up the CO matrix. calculated by hand
# (item, item) -> co_score
CO = {
    (1,2): 3/3,
    (1,3): 1/3,
    (1,4): 0/3,
    (1,5): 0/3,
    (1,6): 1/3,
    (1,7): 0/3,
    (1,8): 0/3,
    (1,9): 1/3,
    (1,10): 2/3,
    
    (2,3): 1/3,
    (2,4): 0/3,
    (2,5): 0/3,
    (2,6): 1/3,
    (2,7): 0/3,
    (2,8): 0/3,
    (2,9): 1/3,
    (2,10): 2/3,
    
    (3,4): 2/3,
    (3,5): 2/3,
    (3,6): 0/3,
    (3,7): 0/3,
    (3,8): 0/3,
    (3,9): 0/3,
    (3,10): 0/3,
    
    (4,5): 1/1,
    (4,6): 0/3,
    (4,7): 0,
    (4,8): 0,
    (4,9): 0,
    (4,10): 0,
    
    (5,6): 0,
    (5,7): 0,
    (5,8): 0,
    (5,9): 0,
    (5,10): 0,
    
    (6,7): 2/3,
    (6,8): 2/3,
    (6,9): 1/1,
    (6,10): 0,
    
    (7,8): 3/3,
    (7,9): 2/3,
    (7,10): 1/3,
    
    (8,9): 2/3,
    (8,10): 1/3,
    (9,10): 0/3
}

def get_CO(i, j):
    if i == j: return 1.0
    i, j = (i,j) if i <= j else (j, i)
    co = CO.get( (i,j) )
    return co

def ICO(c: set):
    if len(c) == 1: return 0.0
    total = 0.0
    for e1, e2 in ((e1,e2) for e1 in c for e2 in c if e1 != e2):
        total += get_CO(e1,e2)
    return total / (len(c)**2 - len(c))

def CCO(c1: set, c2: set):
    total = 0.0
    for e1 in c1:
        for e2 in c2:
            total += get_CO(e1,e2)
    return total / (len(c1)*len(c2))     
```

```python
C = [ c for c in C if len(c) != 0]
print("Initial", C)

def SCCO(c1: set, c2: set):
    scg_score = score_scg(c1.union(c2))
    cco_score = CCO(c1,c2)
    print("habababa", c1,c2,cco_score, scg_score)
    if any((scg for x in c1 for scg in g_map.get(x, []))) \
        and any((scg for x in c2 for scg in g_map.get(x, []))) \
        and scg_score > max(score_scg(c1), score_scg(c2)):
        return scg_score * (cco_score + 1)
    elif cco_score >= max(ICO(c1), ICO(c2)):
        return cco_score
    else: 
        return None #None is -inf

clst = list()
cskip = set()
while True:
    print("loop", C, flush=True)
    changes = False
    cskip.clear()
    clst.clear()
    for i, c1 in enumerate(C):
        if len(c1) == 0 or i in cskip: continue
        best, bscore = None, None
        for j, c2 in ( (j, c2) for j,c2 in enumerate(C) if i < j and len(c2) != 0):
            if j in cskip: continue
            score = SCCO(c1,c2)
            print(i, j, score)
            if score is not None and (bscore is None or score > bscore):
                best, score = c2, score
        #end loop
        if bscore is None and score_scg(c1) < 0:
            changes = True
            clst.extend(set([x]) for x in c1)
        elif best is not None:
            changes = True
            clst.append(set.union(c1,c2))
            cskip.add(i)
            cskip.add(j)
        else:
            clst.append(c1)
            cskip.add(i)
    
    C = [x for x in clst]
    if not changes: 
        break
    
print("DONE", C)
for c in C:
    print("eval: ", c, score_scg(c))
```
